```python
import string
import random

## It corresponds to. length of each solution (chromosome)
POP_SIZE = 1000
SOL_SIZE = 10
USERNAME = 'ec22103'
CROSSOVER_RATE = 0.5
MUTATION_RATE = 0.1

## Here we define the alphabet of our solution
gene_alphabet = list(string.ascii_uppercase) + list(string.digits) + ['_']
print(gene_alphabet)
```

```python
## Generate one random solution (generate parent)
def generate_parent(length, alphabet):
    chromosome = random.choices(alphabet, k=length)
    return ''.join(chromosome)
    
print(generate_parent(SOL_SIZE, gene_alphabet))
```

```python
## Generate population here
def generate_population(population_num, solution_size, alphabet):
    return [generate_parent(solution_size, alphabet) for _ in range(population_num)]


## Here we define the alphabet of our solution
# gene_alphapet = list(string.ascii_uppercase) + list(string.digits) + ['_']
print(generate_population(100, 10, gene_alphabet))
```

```python
from hints import closeness

## guess --> list of solutions
def fitnes_function(guess, username):
    return closeness(guess, username)


# gene_alphapet = list(string.ascii_uppercase) + list(string.digits) + ['_']
# print(gene_alphapet)
guess_list = generate_population(100, SOL_SIZE, gene_alphabet)
print(fitnes_function(guess_list, USERNAME))
    
```

```python
## This function select the best parent for doing reproduction (Mating Selection)
def select_parent(parent_number, fittness_population):
    sorted_dict = {key:value for key, value in sorted(fittness_population.items(), key=lambda item: item[1], reverse=True)}
    return [key for key in sorted_dict.keys()][:parent_number]

## This code creates a population of 100, then return the 20 best parents
guess_list = generate_population(100, SOL_SIZE, gene_alphabet)
fitness_population = fitnes_function(guess_list, USERNAME)
sp = select_parent(20, fitness_population)
print(sp)
```

```python
import numpy as np

## We define the crossover operator to generate new offsprings from parents below
def crossover(parent1, parent2):
    ## Define the cross point randomly
    cross_point = np.random.randint(0, len(parent1))
    # print(parent1)
    # print(parent2)
    ## Swapping
    offspring1 = parent1[:cross_point] + parent2[cross_point:]
    offspring2 = parent2[:cross_point] + parent1[cross_point:]
    return offspring1, offspring2

## Test
crossover(sp[0], sp[1])
```

```python
## Each offspring will be mutated based on a random probability and mutation_rate
def mutate(offspring, mutation_rate, reproduction_num):
    mu_prob = np.random.rand(1)
    if mu_prob <= mutation_rate:
        ## Increment number of reproductions
        reproduction_num[0] += 1
        offspring_char = list(offspring)
        # print('this is original one: {}'.format(offspring))
        ## We do 'Random-Resetting' mutation here
        ## First we choose number of char that we want to change randomly:
        mc_num = np.random.randint(SOL_SIZE)
        ## Index of char that we want tu mutate
        mc_idx = np.random.randint(SOL_SIZE, size=mc_num)
        ## Perform mutation
        for idx in mc_idx:
            offspring_char[idx] = ''.join(random.choices(gene_alphabet, k=1)) ## This is for converting ['A'] to 'a'
        return ''.join(offspring_char)
    else:
        return offspring

## Test mutation
rn = 0
mutate('0X5Q3IEAJA', 0.5, rn)  
        
```

```python

## We have list of selected partents after picking K best parents in a population, now here we aim to generate
## offsprings from this selected parents
## The output is a dict of new children with their fittness values
def create_offspring(parents, crossover_rate, mutation_rate, username, pop_size, reproduction_num):
    ## Based on crossover rate, we decide wheather to create new offspring from the parent, or consider the 
    ## offspring as same its parent
    offsprings = []
    ## So we will have pop_size * (parent1, parent2) after this:
    combination_parent = [random.choices(parents, k=2) for _ in range(pop_size)]
    for parent1, parent2 in combination_parent:
        ## We consider the probability of crossover first:
        co_prob = np.random.rand(1)
        if co_prob <= crossover_rate:
            ## Increment number of reproductions
            reproduction_num[0] += 1
            ## Create offspring with Crossover
            offspring1, offspring2 = crossover(parent1, parent2)
            ## Add them to the the final list of offsprings
            offsprings.extend([offspring1, offspring2])
        else:
            ## offspring will be as same as their parent:
            offsprings.extend([parent1, parent2])
            
    ## Now its time to mutate
    mutated_offsprings = [mutate(offspring, mutation_rate, reproduction_num) for offspring in offsprings]
    
    ## Now we return dict of new children with their fitness values:
    return fitnes_function(mutated_offsprings, username)
    
```

```python
import sys
from timeit import default_timer as timer


def genetic_algorithm(pop_size, sol_size, username, crossover_rate, mutation_rate, parent_num, par_child_ratio, reproduction_num):
    
    start = timer()
    best_fitness = 0
    best_sol = ""
     ## Here we define the alphabet of our solution
    gene_alphapet = list(string.ascii_uppercase) + list(string.digits) + ['_']
    
    ## No we initialize parents for the first time parents (population)
    pop = generate_population(pop_size, sol_size, gene_alphabet)
    # print('This is first population with length of {}'.format(len(pop)))
    # print(pop)
    
    while best_fitness != 1.0:
        ## Now we compute the fitness for all parents and select the best ones to the size of parent_num
        fitness_population = fitnes_function(pop, USERNAME)
        sp = select_parent(parent_num, fitness_population)

        ## Now we do reproduction and generate e new offsprings
        offsp_population = create_offspring(sp, crossover_rate, mutation_rate, username, pop_size, reproduction_num)
        ## Update best fitness
        best_sol = max(offsp_population, key=offsp_population.get)
        best_fitness = offsp_population[best_sol]
        sys.stdout.write('Fitness Score: %f%%          Best Solution: %s  \r' % (best_fitness, best_sol))
        sys.stdout.flush()

        ## Now we select best of 0.75*offsprings + 0.25*parents
        pop = select_parent(int(par_child_ratio*pop_size), fitness_population) + select_parent(int((1-par_child_ratio)*pop_size), offsp_population)
        # print('This is new generation with length of {}:'.format(len(new_gen)))
        # print(new_gen)
    end = timer()
    print('\nThe Secret Phrase is: {}'.format(best_sol))
    # print('The elapsed time is: {} (s)'.format(end-start))
    return best_sol, (end-start)
    # print('Number of reproduction is: {}'.format(reproduction_num[0]))

    
    
```

```python
import itertools

## Size of population
POP_SIZE = 100

## Length of solution
SOL_SIZE = 10

## Username
USERNAME = 'ec22103'

# CROSSOVER_RATE = 0.9
# MUTATION_RATE = 0.1

## Number of selected parents for reproduction
PARENT_NUM = 30

## Ratio of parents that we keep them in each generation (Elitism)
PAR_CHILD_RATIO = 1/10

## Number of reproduction
REP_NUM = [0]

mutation_rates = [0.02, 0.05, 0.1, 0.2, 0.5]
crossover_rates = [0.25, 0.5, 0.75, 0.9, 1]

grid = itertools.product(mutation_rates,crossover_rates)
result_dict = {}

for mut_rate, cross_rate in grid:
    exec_time_list = []
    
    ## perform each combination for 5, then take average
    for i in range(5):
        print('Iter {}: ...'.format(i))
        secret_phase, exec_time = genetic_algorithm(POP_SIZE, SOL_SIZE, USERNAME, cross_rate, mut_rate, PARENT_NUM, PAR_CHILD_RATIO, REP_NUM)
        exec_time_list.append(exec_time)
        print('Elapsed time for crossover_rate = {}, and mutation_rate = {} is: {}'.format(cross_rate, mut_rate, exec_time))
    
    result_dict[(cross_rate, mut_rate)] = np.mean(exec_time_list)
    print('---------------------------------------------')
        
print('---------------------------------------------')
print('---------------------------------------------')
print('The final result for different crossover and mutation rates is:')
print(result_dict)
```
